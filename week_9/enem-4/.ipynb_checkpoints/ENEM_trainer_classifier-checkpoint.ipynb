```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
```

# Getting the data 

```python
original_train_df = pd.read_csv("train.csv")
train_df = original_train_df.copy()
```

# Training using a Random Forest Classifier

### Since the features "TP_ST_CONCLUSAO", "NU_IDADE", "TP_ANO_CONCLUIU" were the most correlated with the target feature ("IN_TREINEIRO"), only these three will be used to train the classifier

```python
features = ["TP_ST_CONCLUSAO", "NU_IDADE", "TP_ANO_CONCLUIU"]
```

```python
x_train = train_df[features]
y_train = train_df["IN_TREINEIRO"]
```

```python
classifier = RandomForestClassifier()
classifier.fit(x_train, y_train)
```

# Getting the test data

```python
original_test_df = pd.read_csv("test.csv")
test_df = original_test_df.copy()
```

# Using the classifier with the test data

```python
x_test = test_df[features]
```

```python
predictions = classifier.predict(x_test)
```

```python
test_predictions = pd.DataFrame({"NU_INSCRICAO": test_df["NU_INSCRICAO"], "IN_TREINEIRO": predictions})
```

```python
test_predictions.to_csv("answer.csv")
```
